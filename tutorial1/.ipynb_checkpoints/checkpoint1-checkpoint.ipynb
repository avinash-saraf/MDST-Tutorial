<h2> Data Cleanup </h2>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Suppress Pandas SettingWithCopyWarning
pd.options.mode.chained_assignment = None
```

Load in the "states_edu.csv" dataset and take a look at the head of the data

```python
df = pd.read_csv("../data/states_edu.csv")
```

You should always familiarize yourself with what each column in the dataframe represents. Read about the states_edu dataset here: https://www.kaggle.com/noriuk/us-education-datasets-unification-project

Use this space to rename columns, deal with missing data, etc. _(optional)_

```python
# Since I've chosen to focus on Grade 4 Math scores, dropping rows with null values for this column 
df.dropna(subset=['AVG_MATH_8_SCORE'], inplace = True)
```

```python
df.isna().sum()
```

```python
# Total revenue and enrollment data are missing for many rows
print("% of rows missing enroll data: ", df["ENROLL"].isna().sum() / df.shape[0] * 100)

print("% of rows missing grades_all_g data : ", df["GRADES_ALL_G"].isna().sum() / df.shape[0] * 100)
```

```python
((df["ENROLL"] - df["GRADES_ALL_G"]) / df["ENROLL"]).describe()
# We can see that the difference between estimtated enrollment and actual enrollment is almost zero
# We can use GRADES_ALL_G data to fill in ENROLL data
```

```python
df["ENROLL"] = df["GRADES_ALL_G"].fillna(df["ENROLL"])
```

<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable.

```python
# let's create a new column which represents expenditure per student
df['SUPPORT_SERVICES_EXPENDITURE_PER_STUDENT'] = df['SUPPORT_SERVICES_EXPENDITURE'] / df['ENROLL']
```

```python
# Filling support_services_expenditure_per_student based on the mean of the ** expenditure per student ** 
# There doesn't seem to be any outliers in the data so i will fill in the missing values using mean not median
df["SUPPORT_SERVICES_EXPENDITURE_PER_STUDENT"].fillna(df["SUPPORT_SERVICES_EXPENDITURE_PER_STUDENT"].mean(), inplace=True)
```

```python
# Will be performing data standardization in the data creation section using a pipeline to get accurate results
```

Feature engineering justification: **Created a new column to show expenditure per student as it would be a good predictor of how well a student does, and filled missing data with the mean of the expediture per student**

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['GRADES_4_G','SUPPORT_SERVICES_EXPENDITURE_PER_STUDENT','AVG_READING_4_SCORE', 'AVG_MATH_8_SCORE', 'AVG_READING_8_SCORE']].dropna()
y = df.loc[X.index]['AVG_MATH_4_SCORE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=.3, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
model = make_pipeline(StandardScaler(), LinearRegression())
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
# root mean squared error -- penalizes large errors
np.mean((model.predict(X_test)-y_test)**2)**0.5
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'AVG_MATH_8_SCORE'

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel('Math 4 Score')
plt.title("Model Behavior On Training Set")
```

```python
col_name = "AVG_MATH_8_SCORE"

f = plt.figure(figsize=(12,6))
plt.scatter(X_test[col_name], y_test, color = "blue")
plt.scatter(X_test[col_name], model.predict(X_test), color = "black")

plt.legend(['True testing','Predicted testing'])
plt.xlabel(col_name)
plt.ylabel('Math 4 Score')
plt.title("Model Behavior on Testing Set")
```
